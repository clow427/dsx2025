```python
import pandas as pd
```

```python
df = pd.read_excel("./RestaurantsAndFoodServices.xlsx", sheet_name=1)
df.head()
```

```python
df.info()
```

```python
df = df[df["State"] == "MA"]
df
```

```python
# Replace the original address column with formatted address: 'address, city MA'
df["Address"] = df["Address"] + ", " + df["City"] + " MA"
df
```

```python
# Drop rows where the Address column is null
df = df.dropna(subset=["Address"])
df
```

```python
# Export the DataFrame to a CSV file
# This will save the filtered DataFrame to 'filtered_restaurants.csv' in the current directory
df.to_csv("filtered_restaurants.csv", index=False)
"CSV export complete."
```

```python
df2 = pd.read_csv("./filtered_restaurants.csv")
df2.head()
```

```python
df2 = df2.dropna(subset=["LongLat"])
df2
```

```python
df2[['Longitude', 'Latitude']] = df2['LongLat'].str.split(',', expand=True)
df2 = df2.drop(columns=['LongLat'])
df2.head()
```

```python
# Set index column for df2
df2 = df2.reset_index(drop=True)

# Add average excess food waste column as the mean of High and Low estimates
high_col = 'Excess Food Estimate, High (tons per year)'
low_col = 'Excess Food Estimate, Low (tons per year)'
df2['AvgExcessFoodWaste'] = df2[[high_col, low_col]].mean(axis=1)

df2.head()
```

```python
df2.to_csv("filtered_restaurants_and_food.csv", index=False)
```
